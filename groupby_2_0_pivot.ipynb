```
def mean_eur2usd(x):
    y = np.mean(x)*1.12
    return y
```

```
# Si le paso un numero a mi función me devuelve simplemente ese numero x 1.12
mean_eur2usd(100)
```

```
# Pero si le paso una lista (Como haría con una columna de un dataframe), entonces
# me calcula primero el promedio y luego lo multiplica por 1.12
mean_eur2usd([100,102])
```

```
# A mi agrupamiento le aplico la función definida previamente para verla en dolares
df.groupby('sex')[['total_bill','prct_tip']].apply(mean_eur2usd)
# Es una variante al .aggregate pero valida para un único indice entre mis columnas
```

```
# Püedo establecer multiples indices en mis filas de la siguiente forma
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(mean_eur2usd)
```

```
# Puedo usar el .apply tmb con funciones definidas por numpy
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(np.std)
```

```
# Podemos tmb definir nuestras propias funciones en el .apply mismo con las funciones lambda
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(lambda x: np.mean(x)*1.12)
```

```
# Para aplicar mas de una función tengo que usar la función .aggregate
df.groupby(['sex','time'])[['total_bill','prct_tip']].aggregate([np.mean,np.max])
```

```
# Agg es una forma resumida de aggregate
df.groupby(['sex','time'])[['total_bill','prct_tip']].agg([np.mean,np.max])
```

```
# Con el uso de un diccionario voy a poder especificar que funciones aplicarque a que columnas.
dict_agg = {'tip':[min, max],'total_bill':[np.mean, mean_eur2usd]}
dict_agg
# Primero armo mi diccionario donde especifico como llaves las columnas y como values mis funciones a aplicar
```

```
# Luego aplico mi diccionario como funcion dentro del .agg
df.groupby(['sex','time'])[['total_bill','tip']].agg(dict_agg)
```

```
# Si mis columnas están definidas en mi diccionario podría tmb obviarlas en la definición de .agg
df.groupby(['sex','time']).agg(dict_agg)
```

```
# Tmb puedo definir filtro para mi dataframe agrupado
def f_filter(x):
    return mean_eur2usd(x['total_bill'].mean()) > 20
# Primero genero mi función de filtro
# Me debería excluir todas las columnas de mujeres que almuerzan si lo aplico al groupby anterior
```

```
df_filtered = df.groupby(['sex','time']).filter(f_filter)
df_filtered
# Me quedaron 209 filas entre 234 originales.
```

```
df_filtered.groupby(['sex','time']).count()
# Compruebo que efectivamente me los excluyo
```

# Pivot o tablas dinámicas (excel)
**Otra forma de agrupar datos en pandas**

```
df
```

```
df.groupby(['sex','time'])[['total_bill']].mean()
```

```
df_gp = df.groupby(['sex','time'])[['total_bill']].mean().reset_index() # Borro los indices y los vuelvo columnas
df_gp
# Guardo este nuevo dataframe
```

```
# Sobre el voy a aplicar una tabla dinamica
df_gp.pivot_table(values='total_bill', index='sex', columns='time') 
# Declaro mis valores, mis indices o filas y mis columnas
```

![](https://pandas-docs.github.io/pandas-docs-travis/_images/reshaping_pivot.png)

```
df.pivot_table(values='total_bill', index='sex', columns='time') # Por defecto .pivot_table calcula promedio. Pero podría calcular otra cosa si se lo paso como argumento
```

```
# Por ejemplo podría calcular la mediana
df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=np.median)
```

```
# Puedo tmb poner una lista de funciones a calcular en mi tabla dinamica o un diccionario para especificar distintas funciones por columna
df_pivot = df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=[np.median,np.std])
df_pivot
```

```
df_pivot.unstack() # mis columnas pasan a ser filas con .unstack
```

```
df_pivot.unstack().reset_index() # al resetear mi indice obtengo un nuevo dataframe con mi pivot table
```
