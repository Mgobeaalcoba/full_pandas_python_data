```
import pandas as pd
import numpy as np
pd.__version__
```

```
pd.options.display.float_format = '{:,.3f}'.format # Trabajare con 3 decimales y la coma como separador de miles
```

```
import seaborn as sns # Importo seaborn para usar una de sus bases de datos
```

```
df = sns.load_dataset('diamonds')
df
```

```
df.groupby("cut") # Aqui agrupo por tipo de corte a mis datos
```

```
# Y con esto puedo obtener medidas de tendencia central por ejemplo.
df.groupby("cut").mean() # Como el promedio
```

```
df.groupby("cut").median() # La mediana
```

Si no quisiera ver la media o mediana de todo mi dataframe puedo seleccionar solo una parte del mismo agregando que columna quiero ver mi medida de tendencia central luego de declarar mi columna para agrupar de la siguiente forma: 

```
df.groupby("cut")["carat"].median()
```

Puedo no solo ver medidas de tendencia central sino tambien contar los datos que tengo para cada agrupamiento:

```
df.groupby("cut")["carat"].count()
```

```
df.groupby("cut").count()
```

Tmb puedo ver valores maximos o minimos de cada agrupamiento: 

```
df.groupby("cut")["carat"].max()
```

```
df.groupby("cut")["carat"].min()
```

Podemos entender el group by como un objeto que contiene paquetes y sobre esos paquetes podemos usar diferencias funciones como venimos viendo y otras como los bucles. 

```
for key_group, group in df.groupby("cut"):
  grouped_price = group["price"].mean()

  print("Cut: {}, Price U$S{} \n".format(key_group, grouped_price))
```

```
# Puedo hacer un agrupamiento por mas de una columna
df.groupby(["cut","color"])["price"].mean()
```

```
# Ese mismo agrupamiento puedo verlo como dataframe así: 
df.groupby(["cut","color"])["price"].mean().to_frame()
```

```
# Con aggregate puedo ver mas de una función para mi agrupamiento
df.groupby(["cut","color"])["price"].aggregate([min,np.mean,max])
```

```
# Puedo con aggregate tmb sumar funciones creadas por mi: 
def mean_kilo(x):
  return np.mean(x)/1000
```

```
df.groupby(["cut","color"])["price"].aggregate([min,np.mean,max,mean_kilo])
```

```
# Puedo armar un diccionario para generar multiples indices no solo en las filas sino tmb en las columnas
dict_agg = {
    "carat":[min,max],
    "price":[np.mean,mean_kilo]
}
dict_agg
```

```
df.groupby(["cut","color"]).aggregate(dict_agg)
```

```
# Tmb podemos crear filtros.
def f_filter(x):
  return mean_kilo(x["price"])>4
```

```
df.groupby("cut").filter(f_filter)
```

```
df.groupby("cut").filter(f_filter)["cut"].unique()
```

Solo me selecciono los tipos de corte premium y fair que son los mas caros y los que están pasando el filtro que establecí.
